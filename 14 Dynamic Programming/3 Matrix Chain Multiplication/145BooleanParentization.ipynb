```python


class Solution:
    mod = 1003
    def countWays(self, N, S):
        
        rows, cols = (N, N)
        dp = [[[-1 for k in range(2)] for i in range(cols + 1)] for j in range(rows + 1)]
        
        return self.solve(S, 0, N-1, 1, dp)
    
    def solve(self, s, i, j, state, dp):
        # for empty string => can't evaluate to true
        if(i > j):
            return 0
        # for string of len 1 (T/F)=> can evaluate to true/false
        if(i == j):
            if(state and s[i] == 'T'):
                return 1 
            elif(state == 0 and s[i] == 'F'):
                return 1
            else:
                return 0
        
        if(dp[i][j][state] != -1):
            return dp[i][j][state]
            
        ways = 0
        
        for k in range(i+1, j, 2):
            
            left_true = self.solve(s, i, k-1, 1, dp)
            left_false = self.solve(s, i, k-1, 0, dp)
            
            right_true = self.solve(s, k+1, j, 1, dp)
            right_false = self.solve(s, k+1, j, 0, dp)
            
            if(s[k] == "&"):
                if(state):
                    ways += (left_true*right_true)
                else:
                    ways += (left_true*right_false) + (left_false*right_true) + (left_false*right_false)
                    
            elif(s[k] == "|"):
                if(state):
                    ways += (left_true*right_true) + (left_true*right_false) + (left_false*right_true)
                else:
                    ways += (left_false*right_false)
            elif(s[k] == "^"):
                if(state):
                    ways +=  (left_true*right_false) + (left_false*right_true)
                else:
                    ways += (left_true*right_true) + (right_false*left_false)
        dp[i][j][state] = ways % self.mod
        return dp[i][j][state]

#{ 
 # Driver Code Starts
#Initial Template for Python 3

import sys
sys.setrecursionlimit(10**6)

if __name__ == '__main__':
    t = int(input())
    for _ in range(t):
        N = int(input())
        S = input()
        
        ob = Solution()
        print(ob.countWays(N, S))
# } Driver Code Ends
```
