```python
class TreeNode():
    def __init__(self,key):
        self.key = key
        self.left = None
        self.right = None
        self.height = 1

class AVLTree():

    def insert_node(self,root,key):

        if not root: return TreeNode(key)
        elif key < root.key: root.left = self.insert_node(root.left, key)
        else: root.right = self.insert_node(root.right, key)

        root.height = 1 + max(self.getHeight(root.left),self.getHeight(root.right))

        balanceFactor = self.getBalance(root)

        if balanceFactor > 1:
            if key < root.left.key: return self.leftRotate(root)
            else: return self.leftrightRotate(root)

        if balanceFactor < -1:
            if key > root.right.key: return self.rightRotate(root)
            else: return self.rightleftRotate(root)

        return root
        
    def rightRotate(self, z):
        y = z.right
        T2 = y.left
        y.left = z
        z.right = T2
        z.height = 1 + max(self.getHeight(z.left),self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left),self.getHeight(y.right))
        return y

    def leftRotate(self, z):
        y = z.left
        T3 = y.right
        y.right = z
        z.left = T3
        z.height = 1 + max(self.getHeight(z.left),self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left),self.getHeight(y.right))
        return y
    
    def leftrightRotate(self,p):
        pl = p.left
        plr = pl.right
        p.left = plr.right
        pl.right = plr.left
        plr.left = pl
        plr.right = p
        plr.height = 1 + max(self.getHeight(plr.left),self.getHeight(plr.right))
        pl.height = 1 + max(self.getHeight(pl.left),self.getHeight(pl.right))
        p.height = 1 + max(self.getHeight(p.left),self.getHeight(p.right))
        return plr
    
    def rigghtleftRotate(self,p):
        pr = p.right
        prl = pr.left
        p.right = prl.left
        pr.left = prl.right
        prl.right = pr
        prl.left = p
        prl.height = 1 + max(self.getHeight(prl.left),self.getHeight(prl.right))
        pr.height = 1 + max(self.getHeight(pr.left),self.getHeight(pr.right))
        p.height = 1 + max(self.getHeight(p.left),self.getHeight(p.right))
        return prl
    
    def getHeight(self, root):
        if not root: return 0
        return root.height

    def getBalance(self, root):
        if not root: return 0
        return self.getHeight(root.left) - self.getHeight(root.right)

    def preorder(self,root):
        print(root.key,end = " ")
        if root.left: self.preorder(root.left)
        if root.right: self.preorder(root.right)
    
    def inorder(self,root):
        if root.left: self.inorder(root.left)
        print(root.key,end = " ")
        if root.right: self.inorder(root.right)
    
    def postorder(self,root):
        if root.left: self.postorder(root.left)
        if root.right: self.postorder(root.right)
        print(root.key,end = " ")
        
    def find_max(self,current):
        if current.right == None: return current.key
        return self.find_max(current.right)

    def find_min(self,current):
        if current.left == None: return current.key
        return self.find_min(current.left)

myTree = AVLTree()
root = None

# nums = [40,20,50,10,30,60,5,25,36,27]
# nums = [5,3,2,6,8]

nums = [30,20,40,10,25,50,5,28,4]
for num in nums:
    root = myTree.insert_node(root, num)

print("INORDER TRAVERSAL: ")
myTree.inorder(root)
print()
print("PREORDER TRAVERSAL: ")
myTree.preorder(root)
print()
print("POSTORDER TRAVERSAL: ")
myTree.postorder(root)
print()

print("MAXIMUM ELEMENT: ")
print(myTree.find_max(root))
print("MINIMUM ELEMENT: ")
print(myTree.find_min(root))
```
